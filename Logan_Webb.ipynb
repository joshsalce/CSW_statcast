## On Fastballs

```python
webb_fastballs = webb[(webb['pitch_type'] == "SI")|
                      (webb["pitch_type"] == "FF")|
                      (webb["pitch_type"] == "FC")]
fastballs = webb_fastballs.shape[0]
```

```python
called_strikes = webb_fastballs[webb_fastballs['description'] == 'called_strike'].shape[0]
```

```python
round(100*(called_strikes / fastballs), 1) #CStr Fastballs
```

```python
sw_strikes = webb_fastballs[(webb_fastballs['description'].isin(swinging_strikes))].shape[0]
```

```python
round(100*(sw_strikes / fastballs), 1)
```

```python
round(100*((called_strikes + sw_strikes)/fastballs), 1) #Fastballs 
```
