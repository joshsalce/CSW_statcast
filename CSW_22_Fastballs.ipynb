```python
statcast_names = []
ids = []

for i in names:
    fname = i.split(' ', 1)[0]
    lname = i.split(' ', 1)[1]
  
    if ((fname + " " + lname == "Mark Appel") or 
        (fname + " " + lname == "James Naile")):
      ids.append(1)

    else:
      if "." in fname:
        fname = fname.replace('.', '. ', 1)

      if fname + " " + lname == "Matthew Boyd":
        fname = fname.replace("Matthew", "Matt")
      elif fname + " " + lname == "Daniel Camarena":
        fname = fname.replace("Daniel", "Dan")
      elif fname + " " + lname == "JT Chargois":
        fname = fname.replace("JT", 'j. t.')
      elif lname[-4:] == " Jr.":
        lname = lname.replace(" Jr.", "")
      elif lname[-4:] == " Sr.":
        lname = lname.replace(" Sr.", "")
      elif fname + " " + lname == "Chi Chi Gonzalez":
        fname = fname.replace("Chi", "Chi Chi")
        lname = lname.replace("Chi Gonzalez", "Gonzalez")
      #elif fname + " " + lname == "Jonathan Heasley":
      #  fname = fname.replace("Jonathan", "Jon")
      elif fname + " " + lname == "Kwang Hyun Kim":
        fname = fname.replace("Kwang", "Kwang Hyun")
        lname = lname.replace("Hyun Kim", "Kim")
      elif fname + " " + lname == "Hyun Jin Ryu":
        fname = fname.replace("Hyun", "Hyun Jin")
        lname = lname.replace("Jin Ryu", "Ryu")
      elif fname + " " + lname == "Michael King":
        fname = fname.replace("Michael", "Mike")
      elif lname == "DÃ­az":
        lname = lname.replace("DÃ­az", "Díaz")
      elif lname == "GutiÃ©rrez":
        lname = lname.replace("GutiÃ©rrez", "Gutiérrez")
      elif lname == "LÃ³pez":
        lname = lname.replace("LÃ³pez", "López")
      elif fname + " " + lname == "AdriÃ¡n MorejÃ³n":
        fname = fname.replace("AdriÃ¡n","Adrián")
        lname = lname.replace("MorejÃ³n","Morejón")
      elif lname == "NÃºÃ±ez":
        lname = lname.replace("NÃºÃ±ez", "Núñez")
      elif fname + " " + lname == "FÃ©lix PeÃ±a":
        fname = fname.replace("FÃ©lix","Félix")
        lname = lname.replace("PeÃ±a","Peña")
      elif fname + " " + lname == "FÃ©lix Bautista":
        fname = fname.replace("FÃ©lix","Félix")
      elif lname == "PÃ©rez":
        lname = lname.replace("PÃ©rez", "Pérez")
      elif lname == "RodrÃ­guez":
        lname = lname.replace("RodrÃ­guez", "Rodríguez")


      if ((fname + " " + lname == "Austin Davis") or 
      (fname + " " + lname == "Luis Castillo") or 
      (fname + " " + lname == "Charlie Morton") or 
      (fname + " " + lname == "Lance McCullers") or
      (fname + " " + lname == "Paul Campbell") or
      (fname + " " + lname == "Adam Eaton") or
      (fname + " " + lname == "Joe Smith") or
      (fname + " " + lname == "Joe Kelly") or 
      (fname + " " + lname == "Erasmo Ramirez") or
      (fname + " " + lname == "Carlos Martinez") or
      (fname + " " + lname == "Drew Anderson")):
        playerid = playerid_lookup(lname, fname)['key_mlbam'][1]
        ids.append(playerid)
      elif fname + " " + lname == "Luis Garcia":
        statcast_names.append(fname + " " + lname)
        garcia1 = playerid_lookup(lname, fname)['key_mlbam'][4]
        garcia2 = playerid_lookup(lname, fname)['key_mlbam'][3]
        ids.append(garcia1)
        ids.append(garcia2)
      elif fname + " " + lname == "Javy Guerra":
        guerra1 = playerid_lookup(lname, fname)['key_mlbam'][0]
        ids.append(guerra1)
      elif fname + " " + lname == "Diego Castillo":
        statcast_names.append(fname + " " + lname)
        castillo1 = playerid_lookup(lname, fname)['key_mlbam'][1]
        castillo2 = playerid_lookup(lname, fname)['key_mlbam'][0]
        ids.append(castillo1)
        ids.append(castillo2)
      else:
        playerid = playerid_lookup(lname, fname)['key_mlbam'][0]
        ids.append(playerid)

    print(fname, lname)
    statcast_names.append(fname + " " + lname)
```

```python
csw_fastballs = []
counter = 1
for i in ids:

    if i != 1:
      statcast_data = statcast_pitcher('2022-04-08', '2022-10-03', player_id = i)
      
      all_fastballs = statcast_data[(statcast_data['pitch_type'] == "SI")|
                        (statcast_data["pitch_type"] == "FF")|
                        (statcast_data["pitch_type"] == "FC")]

      fastballs = all_fastballs.shape[0]
      
      swinging_strikes = ['swinging_strike_blocked','swinging_strike'] 

      called_strikes = all_fastballs[all_fastballs['description'] == 'called_strike'].shape[0]
      sw_strikes = all_fastballs[(all_fastballs['description'].isin(swinging_strikes))].shape[0]
      
      if fastballs == 0:
        csw_per = 0
      else:
        csw_per = round(100*((called_strikes + sw_strikes)/fastballs), 1)

      csw_fastballs.append(csw_per)

    else:
      csw_fastballs.append(0)

    print(counter)
    counter += 1
```

```python
df = pd.DataFrame({'Name': statcast_names,
                  'PlayerID': ids,
                  'ERA': pitching22['ERA'],
                  'IP': pitching22['IP'],
                  'CSW%': csw_fastballs})
df
```
